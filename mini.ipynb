# Student Data Analysis Project

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/student_performance.csv')

print('Dataset Overview:')
print(f"Total number of students: {len(df)}")
print(f"Total number of columns: {len(df.columns)}")
print(f"Number of unique departments: {df['department'].nunique()}")
```

## 3. Department Wise Analysis

```python
print('Department-wise student count:')
print(df['department'].value_counts())
```

## 4. Score Analysis

```python
print('Score Analysis:')
print(f"Highest Math Score: {df['math_score'].max()}")
print(f"Lowest Math Score: {df['math_score'].min()}")
print(f"Average Math Score: {df['math_score'].mean():.2f}")
print(f"Average Programming Score: {df['programming_score'].mean():.2f}")
print(f"Average Attendance Percentage: {df['attendance_percentage'].mean():.2f}")
```

## 5. Gender wise analysis

```python
# Calculate total score for each student
df['total_score'] = df['math_score'] + df['programming_score']

print('Average total score grouped by gender:')
print(df.groupby('gender')['total_score'].mean().round(2))
```

## 6. Top Performers

```python
print('Top Performers (Name, Department, Total Score):')
top_performers = df.sort_values(by='total_score', ascending=False)[['name', 'department', 'total_score']]
print(top_performers.head()) # Display top few, or adjust as needed for all
```

## 7. Attendance Analysis

```python
print('Attendance Analysis (Name, Department, Attendance Percentage):')
print(df[['name', 'department', 'attendance_percentage']])

if (df['attendance_percentage'] > 75).all():
    print('\nMessage: All students have an attendance percentage above 75%.')
else:
    print('\nMessage: Not all students have an attendance percentage above 75%.')
```
